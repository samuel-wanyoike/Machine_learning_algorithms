```python
import numpy as np

A = np.array([[2, 3], [1, 4]])
det = np.linalg.det(A)
print(f"The determinant of the matrix A is: {det}")

inverse_A = np.linalg.inv(A)
print("The inverse of the matrix A is:")
print(inverse_A)

eigenvalues, eigenvectors = np.linalg.eig(A)
print("The eigenvalues of the matrix A are:", eigenvalues)
print("The eigenvectors of the matrix A are:", eigenvectors)

U, S, Vt = np.linalg.svd(A)
print("The left singular vectors (U) of the matrix A are:", U)
print("The singular values of the matrix A are:", S)
print("The right singular vectors (Vt) of the matrix A are:", Vt)

#reconstruct the original matrix from its SVD components
Sigma = np.zeros(A.shape)
np.fill_diagonal(Sigma, S)
A_reconstructed = U @ Sigma @ Vt
print("The reconstructed matrix A from its SVD components is:")
print(A_reconstructed)
print(Sigma)
```
