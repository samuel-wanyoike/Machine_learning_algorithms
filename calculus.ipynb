```python
import sympy as sp

x = sp.Symbol('x')
f=x**3 - 5*x + 7
derivative = sp.diff(f, x)
print("derivative", derivative)
```

```python
x,y = sp.symbols('x y')
f = x**2 + y**2
grad_x = sp.diff(f, x)
grad_y = sp.diff(f, y)


print("partial derivative :", grad_x, grad_y)
```

```python
import numpy as np

# Gradient Descent Implementation
def gradient_descent(X, y, theta, learning_rate, iterations):
    m  = len(y)
    for _ in range(iterations):
        predictions = np.dot(X, theta)
        errors = predictions - y
        gradients = (1/m) * np.dot(X.T, errors)
        theta = theta - learning_rate * gradients
    return theta

#sample data
X = np.array([[1, 1], [1, 2], [1, 3]])
y = np.array([2, 2.5 , 3.5])
theta = np.array([0.1, 0.1])
learning_rate = 0.01
iterations = 1000 

# Running Gradient Descent
optimzed_theta = gradient_descent(X, y, theta, learning_rate, iterations)
print("Optimized parameters:", optimzed_theta)
```

```python
#stochastic gradient descent implementation
import numpy as np

#generate sample data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
# Adding bias term
X_b = np.c_[np.ones((100, 1)), X]

#sgd implementation
def stochastic_gradient_descent(X, y, theta, learning_rate, epochs):
    m = len(y)
    for epoch in range(epochs):
        for i in range(m):
            random_index = np.random.randint(m)
            xi = X[random_index:random_index+1]
            yi = y[random_index:random_index+1]
            prediction = np.dot(xi, theta)
            error = prediction - yi
            gradients = 2 * xi.T.dot(error)
            theta = theta - learning_rate * gradients
    return theta
# Initial parameters
theta = np.random.randn(2,1)
learning_rate = 0.01
epochs = 50

# Running SGD
optimized_theta_sgd = stochastic_gradient_descent(X_b, y, theta, learning_rate, epochs)
print("Optimized parameters (SGD):", optimized_theta_sgd) 
```

```python
def bayes_theorem (prior, likelihood, evidence):
     posterior = (likelihood * prior) / evidence
     return posterior
```

```python
from statistics import mean, variance

data = [2, 4, 6, 8, 10,11]
mean(data)

median = np.median(data)
print("Median:", median)
var = variance(data)
print("Variance:", var)
std_dev = np.std(data)
print("Standard Deviation:", std_dev)
```

```python
from scipy import stats

group1 = [2.1, 2.5, 3.6, 3.9, 4.0]
group2 = [3.2, 3.8, 4.1, 4.5, 5.0]

t_statistic, p_value = stats.ttest_ind(group1, group2)
print("t-statistic:", t_statistic)
print("p-value:", p_value)

alplha = 0.05
if p_value < alplha:
    print("Reject the null hypothesis: Significant difference between groups")
else:
    print("Fail to reject the null hypothesis: No significant difference between groups")
```

```python
import numpy as np
from scipy import stats

data = np.random.normal(loc=50, scale=10, size=1000)

#sample statistics
mean = np.mean(data)
std = np.std(data , ddof=1)
bn = len(data)

#95% confidence interval
z_score = stats.norm.ppf(0.975)
margin_of_error = z_score * (std / np.sqrt(bn))
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
print("95% Confidence Interval:", confidence_interval)
```
